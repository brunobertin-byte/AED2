```python
#pj 1
informacao = {}

def adicionar_contato(nome, numero, email):
    informacao[nome] = {"numero": numero, "email":email}
    print(f"{nome} adicionado com sucesso")


def buscar_contato(nome):
    if nome in informacao:
        print(f"informações do {nome}")
        print(f"numero: {informacao[nome]["numero"]}")
        print(f"E-mail: {informacao[nome]["email"]}")
    else:
        print(f"Contato {nome} não encontrado")


def Listar_contatos():
    print("\nLista de contatos:")
    for nome, info in informacao.items():
        print(f"nome : {nome} , Numero: {info["numero"]} , E-mail: {info["email"]}")
```

```python
adicionar_contato("Sabrina", "1016-2020", "sabrina@email.com")
adicionar_contato("Rosa", "1811-1969", "rosa@email.com")
buscar_contato("Sabrina")
Listar_contatos()
```

```python
#pj 2
tarefas = ["Lavar roupa", "Estudar", "trabalhar"]

def adicionar_terefa(tarefa):
    tarefas.append(tarefa)
    print(f"Tarefa '{tarefa}' adicionada")

def marcar_concluida(tarefa):
    if tarefa in tarefas:
        tarefas.remove(tarefa)
        print(f"Tarefa '{tarefa}' marcada como concluida")
    else:
        print(f"Tarefa '{tarefa}' não encontrada")

def vizualizar_tarefas():
    print("\n Lista de Tarefas:")
    for tarefa in tarefas:
        print(f"-{tarefa}")
```

```python
adicionar_terefa("Lavar louças")
marcar_concluida("Lavar roupa")
vizualizar_tarefas()
```
