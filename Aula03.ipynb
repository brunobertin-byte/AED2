```python
# função que realiza a busca 
def busca_linear(vetor, valor_procurado):
    
    # percorre todo o vetor
    for i in range(len(vetor)):
        
        # verifica se o elemento atual é o que estamos procurando
        if vetor[i] == valor_procurado:
            return i  # retorna a posição onde encontrou
    
    # se terminar o loop e não encontrar
    return -1


# vetor (lista) não ordenado
vetor = [23, 5, 89, 12, 7, 34]

# valor que queremos buscar
valor = 12

# chama a função
resultado = busca_linear(vetor, valor)

# verifica se encontrou
if resultado != -1:
    print("Valor encontrado na posição:", resultado)
else:
    print("Valor não encontrado")
```

```python
def buscar_cliente(nome, lista_clientes):
    for i in range(len(lista_clientes)):
        if lista_clientes[i] == nome:
            return i  # encontrou, retorna a posição
    return -1  # não encontrou, retorna -1


# Lista de exemplo e testes da função
clientes = ["Ana", "Bruno", "Carlos", "Daniel", "Elisa"]

busca1 = "Carlos"
busca2 = "Sabrina"

pos1 = buscar_cliente(busca1, clientes)
if pos1 != -1:
    print(f"Cliente '{busca1}' encontrado na posição {pos1}.")
else:
    print(f"Cliente '{busca1}' não encontrado na lista.")

pos2 = buscar_cliente(busca2, clientes)
if pos2 != -1:
    print(f"Cliente '{busca2}' encontrado na posição {pos2}.")
else:
    print(f"Cliente '{busca2}' não encontrado na lista.")
```
