```python
def pesquisa_binaria(lista, alvo):
    inicio = 0
    fim = len(lista) - 1

    while inicio <= fim:
        meio = (inicio + fim) // 2

        if lista[meio] == alvo:
            return meio
        elif lista[meio] < alvo:
            inicio = meio + 1
        else:
            fim = meio - 1

    return -1


numeros = [3, 7, 12, 18, 21, 25, 30]

resultado = pesquisa_binaria(numeros, 21)

print(resultado)
```

```python
# Busca Sequencial
def busca_sequencial(lista, alvo):
    for i in range(len(lista)):
        if lista[i] == alvo:
            return i
    return -1


# Busca Binária
def busca_binaria(lista, alvo):
    inicio = 0
    fim = len(lista) - 1

    while inicio <= fim:
        meio = (inicio + fim) // 2

        if lista[meio] == alvo:
            return meio
        elif lista[meio] < alvo:
            inicio = meio + 1
        else:
            fim = meio - 1

    return -1


# Lista ordenada
numeros = [3, 7, 12, 18, 21, 25, 30, 35, 40]

valor = 25

print("Busca Sequencial:", busca_sequencial(numeros, valor))
print("Busca Binária:", busca_binaria(numeros, valor))
```

```python
produtos = [
    {"nome": "Cadeira Gamer", "preco": 1200, "estoque": 5},
    {"nome": "Fone Bluetooth", "preco": 250, "estoque": 15},
    {"nome": "Mouse Gamer", "preco": 150, "estoque": 30},
    {"nome": "Notebook Dell", "preco": 4500, "estoque": 3},
    {"nome": "Teclado Mecânico", "preco": 350, "estoque": 10}
]

def busca_produto(lista, nome):

    inicio = 0
    fim = len(lista) - 1

    while inicio <= fim:

        meio = (inicio + fim) // 2
        nome_meio = lista[meio]["nome"]

        if nome_meio == nome:
            return lista[meio]

        elif nome_meio < nome:
            inicio = meio + 1

        else:
            fim = meio - 1

    return None
    
produto_procurado = input("Digite o nome do produto: ")

resultado = busca_produto(produtos, produto_procurado)

if resultado:
    print("Produto encontrado!")
    print("Nome:", resultado["nome"])
    print("Preço: R$", resultado["preco"])
    print("Estoque:", resultado["estoque"])
else:
    print("Produto não encontrado.")
```
